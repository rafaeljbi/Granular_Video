```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_fscore_support, confusion_matrix
import os
import re
import seaborn as sns
```

```python
# --- Configuração ---

# 1. Mapeamento dos pares de arquivos e a classe-alvo de cada um
#    (Ajuste os nomes e os caminhos dos arquivos)
FILE_PAIRS = {
    "Smartwatch": {
        "log": "D:/Tese_UPM/Granular_Recognition/Projetos/Granular_Video/Comparison/Comparison_files/log_Smartwatch.txt",
        "gt": "D:/Tese_UPM/Granular_Recognition/Projetos/Granular_Video/Comparison/Comparison_files/Ground_truth_Smartwatch.txt",
        "target_class": "smartwatch"
    },
    "Blender": {
        "log": "D:/Tese_UPM/Granular_Recognition/Projetos/Granular_Video/Comparison/Comparison_files/log_Blender.txt",
        "gt": "D:/Tese_UPM/Granular_Recognition/Projetos/Granular_Video/Comparison/Comparison_files/Ground_truth_Blender.txt",
        "target_class": "blender"
    },
    "Hair Dryer": {
        "log": "D:/Tese_UPM/Granular_Recognition/Projetos/Granular_Video/Comparison/Comparison_files/log_Hair_dryer.txt", # Corrigido para "Hainr_dryer"
        "gt": "D:/Tese_UPM/Granular_Recognition/Projetos/Granular_Video/Comparison/Comparison_files/Ground_truth_Hair_dryer.txt",
        "target_class": "hair_dryer"
    },
    "Mixer": {
        "log": "D:/Tese_UPM/Granular_Recognition/Projetos/Granular_Video/Comparison/Comparison_files/log_Mixer.txt",
        "gt": "D:/Tese_UPM/Granular_Recognition/Projetos/Granular_Video/Comparison/Comparison_files/Ground_truth_Mixer.txt",
        "target_class": "mixer"
    },
    "Grain Grinder": {
        "log": "D:/Tese_UPM/Granular_Recognition/Projetos/Granular_Video/Comparison/Comparison_files/log_Grain_Grinder.txt",
        "gt": "D:/Tese_UPM/Granular_Recognition/Projetos/Granular_Video/Comparison/Comparison_files/Ground_truth_Blender_Grain_grinder.txt", # Mapeado conforme sua lista
        "target_class": "grain_grinder"
    }
}

MAX_TIME_SECONDS = 120
FPS = 30.0
BIN_SIZE = 1.0 / FPS
OUTPUT_FOLDER = f"Resultados_Analise_Frame_a_Frame/"
```

```python
def clean_class_name(name):
    """Limpa e padroniza os nomes das classes (minúsculas, remove espaços/underscores)."""
    if not isinstance(name, str):
        return ""
    return re.sub(r'[\s_]+', '', name.lower())

def load_and_bin_data(log_file, gt_file, target_class):
    """
    Carrega os arquivos de log e ground truth, agrupa por frame (binning)
    e cria a "ficha de pontuação".
    """
    try:
        log_df = pd.read_csv(log_file)
        gt_df = pd.read_csv(gt_file)
    except FileNotFoundError as e:
        print(f"Erro: Arquivo não encontrado: {e.filename}")
        return None, None
    except Exception as e:
        print(f"Um erro inesperado ocorreu ao ler os arquivos: {e}")
        return None, None

    # Limpa nomes das classes
    log_df['Classe_Limpa'] = log_df['Classe'].apply(clean_class_name)
    gt_df['Classe_Limpa'] = gt_df['Classe'].apply(clean_class_name)
    target_class_clean = clean_class_name(target_class)

    # Cria os "baldes" no tamanho de 1 frame
    bins = np.arange(0, MAX_TIME_SECONDS + BIN_SIZE, BIN_SIZE)
    bin_labels = bins[:-1]
    
    # Processa os Logs dos Modelos
    log_df_filtered = log_df[log_df['Classe_Limpa'] == target_class_clean]
    log_df_filtered = log_df_filtered.copy()
    log_df_filtered['interval'] = pd.cut(log_df_filtered['Timestamp_Video_s'], bins=bins, right=False, labels=bin_labels)
    pred_especialista = log_df_filtered[log_df_filtered['Modelo'] == 'Especialista'].groupby('interval', observed=False).size().reindex(bin_labels, fill_value=0) > 0
    pred_geral = log_df_filtered[log_df_filtered['Modelo'] == 'Geral'].groupby('interval', observed=False).size().reindex(bin_labels, fill_value=0) > 0
    
    # Processa o Ground Truth (GT)
    gt_df_filtered = gt_df[
        (gt_df['Classe_Limpa'] == target_class_clean) & 
        (gt_df['Anotador'] == 'Manual') # <-- CORRIGIDO
    ]
    gt_df_filtered = gt_df_filtered.copy()
    gt_df_filtered['interval'] = pd.cut(gt_df_filtered['Timestamp_Video_s'], bins=bins, right=False, labels=bin_labels)
    gt_truth = gt_df_filtered.groupby('interval', observed=False).size().reindex(bin_labels, fill_value=0) > 0
    
    # Monta a "Ficha de Pontuação"
    scorecard = pd.DataFrame({
        'interval': bin_labels,
        'Ground_Truth': gt_truth,
        'Especialista': pred_especialista,
        'Geral': pred_geral
    })
    scorecard = scorecard.set_index('interval')
    
    return scorecard

def plot_timeline(df, title, filename):
    """Cria um gráfico de 'Gantt' para visualização e o exibe."""
    fig, ax = plt.subplots(figsize=(20, 4))
    
    categories = ['Geral', 'Especialista', 'Ground_Truth']
    colors = {'Geral': 'C0', 'Especialista': 'C1', 'Ground_Truth': 'C2'}
    
    for i, cat in enumerate(categories):
        # Lógica do gráfico corrigida
        ranges = []
        start_sec = None
        for sec, value in df[cat].items():
            if value and start_sec is None:
                start_sec = sec 
            elif not value and start_sec is not None:
                duration = sec - start_sec
                ranges.append((start_sec, duration))
                start_sec = None
        
        if start_sec is not None:
            duration = MAX_TIME_SECONDS - start_sec
            ranges.append((start_sec, duration))
            
        if ranges:
            ax.broken_barh(ranges, (i - 0.4, 0.8), facecolors=colors[cat], label=cat)

    ax.set_yticks(range(len(categories)))
    ax.set_yticklabels(categories)
    ax.set_xlabel('Tempo (segundos)')
    ax.set_title(f'Linha do Tempo de Detecção: {title} (Frame-a-Frame)')
    ax.legend(loc='upper right')
    ax.grid(True, axis='x', linestyle='--', alpha=0.6)
    ax.set_xlim(0, MAX_TIME_SECONDS)
    
    plt.tight_layout()
    plt.savefig(filename)
    print(f"  > Gráfico de Timeline salvo em: {filename}")
    plt.show() # <-- MUDANÇA AQUI: Mostra o gráfico no notebook
    # plt.close(fig) # Removido para que o gráfico apareça

# --- NOVA FUNÇÃO ---
def plot_confusion_matrix(cm, title, filename):
    """Cria um heatmap da matriz de confusão, salva e exibe."""
    
    # O cm vem como [[TN, FP], [FN, TP]]
    tn, fp, fn, tp = cm.ravel()
    
    # Cria as anotações
    group_names = ['Verdadeiro Negativo (TN)', 'Falso Positivo (FP)',
                   'Falso Negativo (FN)', 'Verdadeiro Positivo (TP)']
    group_counts = [f"{tn}", f"{fp}", f"{fn}", f"{tp}"]
    group_total = np.sum(cm)
    group_percentages = [f"{v / group_total:.2%}" for v in cm.ravel()]
    
    # Formata as labels
    labels = [f"{v1}\n{v2}\n({v3})" for v1, v2, v3 in zip(group_names, group_counts, group_percentages)]
    labels = np.asarray(labels).reshape(2, 2)
    
    # Cria o plot
    fig, ax = plt.subplots(figsize=(8, 6))
    sns.heatmap(cm, annot=labels, fmt='', cmap='Blues', ax=ax, cbar=False,
                annot_kws={"size": 12, "va": "center"})
    
    ax.set_title(title, fontsize=14)
    ax.set_xlabel('Label Prevista', fontsize=12)
    ax.set_ylabel('Label Verdadeira (Ground Truth)', fontsize=12)
    ax.set_xticklabels(['Negativo (Não Viu)', 'Positivo (Viu)'])
    ax.set_yticklabels(['Negativo (Não Tinha)', 'Positivo (Tinha)'])
    
    plt.tight_layout()
    plt.savefig(filename)
    print(f"  > Matriz de Confusão salva em: {filename}")
    plt.show() # Mostra o gráfico no notebook
    plt.close(fig) # Fecha a figura após salvar e mostrar

def analyze_model(scorecard, model_name):
    """Calcula as métricas e retorna a Matriz de Confusão."""
    y_true = scorecard['Ground_Truth']
    y_pred = scorecard[model_name]
    
    precision, recall, f1, _ = precision_recall_fscore_support(y_true, y_pred, average='binary', zero_division=0)
    
    # --- MUDANÇA AQUI: Garante uma matriz 2x2 ---
    # labels=[False, True] força a ordem [TN, FP, FN, TP]
    cm = confusion_matrix(y_true, y_pred, labels=[False, True])
    tn, fp, fn, tp = cm.ravel()

    return {
        "Model": model_name,
        "TP": tp, "FP": fp, "FN": fn, "TN": tn,
        "Precision": precision, "Recall": recall, "F1-Score": f1,
        "ConfusionMatrix": cm  # <-- MUDANÇA AQUI: Retorna a matriz
    }

# --- Script Principal ---
# Montar o drive (necessário se ainda não o fez)
try:
    from google.colab import drive
    drive.mount('/content/drive')
except ModuleNotFoundError:
    print("Não estou no Colab, pulando a montagem do drive.")


if not os.path.exists(OUTPUT_FOLDER):
    os.makedirs(OUTPUT_FOLDER)
    print(f"Pasta de resultados criada em: {OUTPUT_FOLDER}")
    
print(f"Iniciando análise (Frame-a-Frame @ {FPS} FPS)... Resultados serão salvos em '{OUTPUT_FOLDER}'")

all_results = []

for video_name, files in FILE_PAIRS.items():
    print(f"\n--- Processando: {video_name} ---")
    
    scorecard = load_and_bin_data(files['log'], files['gt'], files['target_class'])
    
    if scorecard is None:
        print("  > Erro ao carregar dados. Pulando...")
        continue
        
    result_especialista = analyze_model(scorecard, 'Especialista')
    result_geral = analyze_model(scorecard, 'Geral')
    
    result_especialista['Video'] = video_name
    result_geral['Video'] = video_name
    all_results.extend([result_especialista, result_geral])
    
    print(f"  Resultados da Análise (por frame):")
    print(f"  Modelo 'Especialista':")
    print(f"    - F1-Score: {result_especialista['F1-Score']:.3f} | Precision: {result_especialista['Precision']:.3f} | Recall: {result_especialista['Recall']:.3f}")
    print(f"    - TP: {result_especialista['TP']} frames | FP: {result_especialista['FP']} frames | FN: {result_especialista['FN']} frames")
    print(f"  Modelo 'Geral':")
    print(f"    - F1-Score: {result_geral['F1-Score']:.3f} | Precision: {result_geral['Precision']:.3f} | Recall: {result_geral['Recall']:.3f}")
    print(f"    - TP: {result_geral['TP']} frames | FP: {result_geral['FP']} frames | FN: {result_geral['FN']} frames")

    # --- NOVO: Plotar Matrizes de Confusão ---
    # Modelo Especialista
    cm_filename_esp = os.path.join(OUTPUT_FOLDER, f"Matriz_Confusao_Especialista_{video_name.replace(' ', '_')}.png")
    cm_title_esp = f"Matriz de Confusão: Modelo 'Especialista'\nVídeo: {video_name}"
    plot_confusion_matrix(result_especialista['ConfusionMatrix'], cm_title_esp, cm_filename_esp)
    
    # Modelo Geral
    cm_filename_ger = os.path.join(OUTPUT_FOLDER, f"Matriz_Confusao_Geral_{video_name.replace(' ', '_')}.png")
    cm_title_ger = f"Matriz de Confusão: Modelo 'Geral'\nVídeo: {video_name}"
    plot_confusion_matrix(result_geral['ConfusionMatrix'], cm_title_ger, cm_filename_ger)
    
    # --- Plotar Linha do Tempo (agora também exibe) ---
    plot_filename = os.path.join(OUTPUT_FOLDER, f"Timeline_{video_name.replace(' ', '_')}.png")
    plot_timeline(scorecard, video_name, plot_filename)
    
# --- Resultados Finais ---
print("\n--- Análise Concluída ---")

results_df = pd.DataFrame(all_results)
cols = ['Video', 'Model', 'F1-Score', 'Precision', 'Recall', 'TP', 'FP', 'FN', 'TN']
results_df = results_df[cols]

csv_filename = os.path.join(OUTPUT_FOLDER, f"Resultados_Comparativos_Finais_Frame_a_Frame.csv")
results_df.to_csv(csv_filename, index=False, float_format='%.4f')

print(f"\nResultados Finais (F1-Score Médio):")
print(results_df.groupby('Model')[['F1-Score', 'Precision', 'Recall']].mean().to_markdown(floatfmt=".3f"))

print(f"\nTabela comparativa completa salva em: {csv_filename}")
```
